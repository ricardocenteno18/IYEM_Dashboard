```python
import pandas as pd
import numpy as np
```

```python
#read a xlsx file
df = pd.read_excel('Seguimiento_Beneficiarias_DE_IYEM.xlsx', sheet_name='Beneficiarios', header = None)
#ignore the first 3 rows
df = df.iloc[5:]
df.header = None
#set the first row as the header
df.columns = df.iloc[0]
#delete the first row
df = df.iloc[1:]
df.head()
```

```python
from dash import dash, dcc, html 
import plotly.express as px
from dash.dependencies import Input, Output
```

```python
#creating a dropdown list
dropdown_list = []
for i in df['MUNICIPIO']:
    dropdown_list.append({'label': i, 'value': i})

#Creating the app
app = dash.Dash()

colors = {
    'background': '#111111',
    'text': '#7FDBFF'
}

#Creating the layout
app.layout = html.Div(style={'backgroundColor': colors['background']}, children=[
    html.H1(
        children = 'Seguimiento de Beneficiarias de IYEM Dashboard',
    style = {'textAlign': 'center',
            'color': colors['text'], 'font-family': 'Helvetica', 'font-size': '40px'}), 
    html.Div(dcc.Dropdown(id='dropdown', style={'backgroundColor': colors['text']}, 
                options=dropdown_list,
                value='Select')),
    html.Div(dcc.Graph(id='map'), style={'color': colors['text']}),
])

#Creating the callback
@app.callback( 
    Output(component_id='map', component_property='figure'),
    Input(component_id='dropdown', component_property='value')
)
def update_graph(value):
    df1 = df[df['MUNICIPIO'] == value]
    fig = px.choropleth(df1, locations='MUNICIPIO',
                        #color='GDP ()',
                        hover_name='MUNICIPIO',
                        color_continuous_scale=px.colors.sequential.Plasma)
    return fig

#Running the app
if __name__ == '__main__':
    app.run_server()
```
